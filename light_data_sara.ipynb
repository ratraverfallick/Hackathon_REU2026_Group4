# 2. Zip Code Extraction

```python
service_light["zip_code"] = service_light["address"].str.extract(r",\s*TX,\s*(\d{5})")
```

```python
service_light["created_date"] = pd.to_datetime(service_light["created_date"])
service_light["closed_date"] = pd.to_datetime(service_light["closed_date"])
```

```python
#only closed requests
closed = service_light[
    service_light["status"].str.contains("Closed", case=False, na=False)
].copy()
```

```python
#service time in days
closed["service_time_days"] = (
    closed["closed_date"] - closed["created_date"]
).dt.total_seconds() / 86400
```

```python
#compare by zip code 
zip_summary = (
    closed.groupby("zip_code")["service_time_days"]
    .agg(["count", "mean", "median", "min", "max"])
    .sort_values("mean", ascending=False)
    .reset_index()
)

zip_summary
```

```python
#bar chart mean
import matplotlib.pyplot as plt

zip_summary_filtered = zip_summary.sort_values("mean", ascending=False)
plt.figure(figsize=(12, 6))
plt.bar(zip_summary_filtered["zip_code"], zip_summary_filtered["mean"])
plt.xticks(rotation=90)
plt.xlabel("ZIP Code")
plt.ylabel("Average Service Time (Days)")
plt.title("Average Street Lighting Service Time by ZIP Code")
plt.show()
```

```python
#bar chart median
zip_summary_filtered = zip_summary.sort_values("median", ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(zip_summary_filtered["zip_code"], zip_summary_filtered["median"])
plt.xticks(rotation=90)
plt.xlabel("ZIP Code")
plt.ylabel("Median Service Time (Days)")
plt.title("Median Street Lighting Service Time by ZIP Code")
plt.show()
```

```python
zip_stats = (
    closed.groupby("zip_code")["service_time_days"]
    .agg(["mean", "std", "count"])
    .reset_index()
)

zip_stats

```

```python
#CI 
zip_stats["se"] = zip_stats["std"] / (zip_stats["count"] ** 0.5)
zip_stats["ci95"] = 1.96 * zip_stats["se"]
```

```python
plt.figure(figsize=(12, 6))

plt.errorbar(
    zip_stats["zip_code"],
    zip_stats["mean"],
    yerr=zip_stats["std"],
    fmt="o",
    capsize=5
)
plt.ylim(bottom=0)
plt.xticks(rotation=90)
plt.xlabel("ZIP Code")
plt.ylabel("Average Service Time (Days)")
plt.title("Average Street Lighting Service Time by ZIP Code")
plt.ylim(0, zip_stats["mean"].max() + zip_stats["ci95"].max() + 5)
plt.show()
```

```python
#standard error instead of standard deviation 
zip_stats["se"] = zip_stats["std"] / zip_stats["count"]**0.5
```

```python
plt.figure(figsize=(12, 6))

plt.errorbar(
    zip_stats["zip_code"],
    zip_stats["mean"],
    yerr=zip_stats["se"],
    fmt="o",
    capsize=5
)

plt.xticks(rotation=90)
plt.xlabel("ZIP Code")
plt.ylabel("Average Service Time (Days)")
plt.title("Average Street Lighting Service Time by ZIP Code (±1 SE)")
plt.show()
```

```python
#Standard deviation (SD): shows how much individual service times vary within each ZIP code.
#Standard error (SE): shows the uncertainty in the estimated mean.
#95% confidence intervals: are often preferred in reports because they show a plausible range for the true average.
```

```python
# Confidence Intervals

# Summary statistics
zip_stats = (
    closed.groupby("zip_code")["service_time_days"]
    .agg(["mean", "std", "count"])
    .reset_index()
)

# Standard error
zip_stats["se"] = zip_stats["std"] / (zip_stats["count"] ** 0.5)

# 95% confidence interval
zip_stats["ci95"] = 1.96 * zip_stats["se"]

# (Optional) Only keep ZIP codes with at least 10 requests
zip_stats = zip_stats[zip_stats["count"] >= 10]

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.errorbar(
    zip_stats["zip_code"].astype(str),
    zip_stats["mean"],
    yerr=zip_stats["ci95"],
    fmt="o",
    capsize=5
)

plt.xticks(rotation=90)
plt.xlabel("ZIP Code")
plt.ylabel("Average Service Time (Days)")
plt.title("Mean Street Lighting Service Time by ZIP Code (95% Confidence Intervals)")
plt.ylim(bottom=0)

plt.show()
```
