# 2. Zip Code Extraction

```python
service_light["zip_code"] = service_light["address"].str.extract(r",\s*TX,\s*(\d{5})")
```

```python
service_light["created_date"] = pd.to_datetime(service_light["created_date"])
service_light["closed_date"] = pd.to_datetime(service_light["closed_date"])
```

```python
#only closed requests
closed = service_light[
    service_light["status"].str.contains("Closed", case=False, na=False)
].copy()
```

```python
#service time in days
closed["service_time_days"] = (
    closed["closed_date"] - closed["created_date"]
).dt.total_seconds() / 86400
```

```python
#compare by zip code 
zip_summary = (
    closed.groupby("zip_code")["service_time_days"]
    .agg(["count", "mean", "median", "min", "max"])
    .sort_values("mean", ascending=False)
    .reset_index()
)

zip_summary
```

```python
#bar chart mean
import matplotlib.pyplot as plt

zip_summary_filtered = zip_summary.sort_values("mean", ascending=False)
plt.figure(figsize=(12, 6))
plt.bar(zip_summary_filtered["zip_code"], zip_summary_filtered["mean"])
plt.xticks(rotation=90)
plt.xlabel("ZIP Code")
plt.ylabel("Average Service Time (Days)")
plt.title("Average Street Lighting Service Time by ZIP Code")
plt.show()
```

```python
#bar chart median
zip_summary_filtered = zip_summary.sort_values("median", ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(zip_summary_filtered["zip_code"], zip_summary_filtered["median"])
plt.xticks(rotation=90)
plt.xlabel("ZIP Code")
plt.ylabel("Median Service Time (Days)")
plt.title("Median Street Lighting Service Time by ZIP Code")
plt.show()
```

# 3. Filtering the Data

```python
service_light['service_request_type'].unique()
```

```python
maintenance = service_light[service_light['service_request_type'] == "Street Lighting - Maintenance- TPW"]
```

```python
maintenance
```

```python
maintenance["created_date"] = pd.to_datetime(maintenance["created_date"])

light_df = maintenance[
    (maintenance["created_date"].dt.year >= 2021) &
    (maintenance["created_date"].dt.year <= 2025)
]

light_df
```

```python
light_df.to_csv("street_lighting_2021_2025.csv", index=False)
```
