```python
from pathlib import Path
import pandas as pd

s_path_data = Path(r"C:\2026_06_26_Hackathon\Data")
s_file = s_path_data / "Dallas_City_Owned_Street_Lights.csv"
s_out_summary = s_path_data / "Dallas_City_Owned_Street_Lights_COLUMN_SUMMARY.csv"

df = pd.read_csv(s_file, dtype=str, low_memory=False)

summary_rows = []
for col in df.columns:
    s = df[col].fillna("").astype(str).str.strip()
    summary_rows.append({
        "column": col,
        "nonblank_count": (s != "").sum(),
        "blank_count": (s == "").sum(),
        "unique_count": s[s != ""].nunique(),
        "example_values": " | ".join(s[s != ""].drop_duplicates().head(8).to_list())
    })

df_summary = pd.DataFrame(summary_rows).sort_values(["nonblank_count", "unique_count"], ascending=[False, False])
df_summary.to_csv(s_out_summary, index=False)

print("Rows:", len(df))
print("Columns:", len(df.columns))
print("Saved:", s_out_summary)
print(df_summary.to_string(index=False))
```
